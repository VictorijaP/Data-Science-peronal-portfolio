Всего было сделано 9835 покупок
Самый большой чек - 32 товара, самый маленький - 1 товар.
Средний чек - 3 - 4 товара.

```python
# заполнение пропусков последним значением в строке

df1 = df.ffill(axis=1)
```

```python
df1.set_index('Item(s)', inplace=True)
```

```python
df1.head()
```

```python
transactions = []
for i in range(0, 9835):
    transactions.append([str(df1.values[i, j]) for j in range(0, 32) ])
```

```python
result = apriori(transactions, min_support=0.0045, min_confidence=0.2, min_lift=3, min_lenght=2)
```

```python
paires = pd.DataFrame(columns=['Base Item', 'Add Item', 'Rule', 'Support', 'Confidence', 'Lift'])

for item in result:
    pair = item[0]
    items = [x for x in pair]
    
    add_item = items[0]
    base_item = items[1]
    
    rules = items[0] + " -> " + items[1]
    paires = paires.append(
                        {
                            'Base Item': str(items[1]),
                            'Add Item': str(items[0]),
                            'Rule': rules, 
                            'Support': str(item[1]), 
                            'Confidence': str(item[2][0][2]), 
                            'Lift': str(item[2][0][3])
                        }, ignore_index=True
                    )
   
```

```python
# вывод пар товаров

paires
```

```python
# Определение самых продаваемых товаров

pairs_sort = paires.sort_values(by=['Support'], ascending=False)
pairs_sort
```

```python
print('Топ-3 продукти, що продаються')
pairs_sort['Base Item'].head(4).unique()
```

```python
# Список пор продаваемых товаров

paires2 = paires.sort_values(by=['Lift'], ascending=False)
paires2
```

```python
print('Топ-2 пары продуктов')
paires2['Rule'].head(2).unique()
```

В исходном наборе данных нет информации о ценах, о дне (праздник, выходной, будний день) и о времени совершения транзакций, о том, были ли в этот день маркетинговые активности или нет.
Я буду исходить из допущения, что данные взяты за обычный будний день, маркетинговые мероприятие (акции, рассылки и т.д.) не проводились какое-то время, т.е. данные "базовые".
Исходя из полученныъ результатов можно сказать, что:
- Для самых продаваемых продуктов можно увеличить маржу и не включать эти продукты в рекламные акции и рассылки
- Самые "сильные пары" продуктов можно использоваь для продвижения нового продукта этой же товарной группы, и еще для оптимального размещения продуктов на полке (если это формат обычного ритейла).
